```python
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from sklearn import datasets, linear_model

%matplotlib inline
```

```python
# read data into a DataFrame
data = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', index_col=0)
data.head()
```

```python
# Set the grid
gs = gridspec.GridSpec(2, 2)

# Set overall plot size
plt.subplots(figsize = (10,10))

# Define subplot axis
ax_tv = plt.subplot(gs[0, :])
ax_radio = plt.subplot(gs[1, 0])
ax_news = plt.subplot(gs[1, 1])

# Make plot
ax_tv.plot(data['TV'], data['Sales'], 'ro', )
ax_radio.plot(data['Radio'], data['Sales'], 'bo')
ax_news.plot(data['Newspaper'], data['Sales'], 'go')
```
